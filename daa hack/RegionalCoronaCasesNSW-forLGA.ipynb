```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
import  pandas as pd
import numpy as np
```

```python
url = 'https://www.health.nsw.gov.au/Infectious/diseases/Pages/covid-19-lga.aspx'

response = requests.get(url)
```

```python
coverpage = response.content
soup1 = BeautifulSoup(coverpage, 'html5lib')
```

```python
heading=soup1.find_all('table',class_="moh-rteTable-6")
```

```python
name=heading[0]
```

```python
name
```

```python
import re

Regions=[]
values=[]
for trs in name.find_all('tr'):
        heading1=trs.get_text(strip=True)
        word1 = " ".join(re.findall("[a-zA-Z]+",heading1 ))
        tds = trs.find_all('td')
        row = [elem.text.strip(u'\u200b') for elem in tds]
        Regions.append(word1)
        values.append(row)
        
       
        
```

```python

Regions=Regions[1:]
Regions=Regions[:-1]
```

```python
Regions
```

```python
values
```

```python
flat_list = [item for sublist in values for item in sublist]

```

```python
flat_list1=[]
for i in flat_list:
    flat_list1.append(i.strip())
```

```python
flat_list1=flat_list1[1::2]
```

```python
flat_list1=flat_list1[:-1]
```

```python
flat_list
newflatlist=[]
for x in flat_list1:
    if "-" in x:
        newx=(x.split("-"))
        newflatlist.append(newx[1])
    else:
        newflatlist.append(x)
```

```python
import pandas as pd
Regional_NSW_Total = {'Region':Regions,
        'cases': newflatlist
        }
```

```python
df = pd.DataFrame(Regional_NSW_Total, columns = ['Region', 'cases'])
df["State"]="New South Wales"
```
