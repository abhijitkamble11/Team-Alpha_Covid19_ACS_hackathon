```python
from pandas import DataFrame as df
from math import sin, cos, sqrt, atan2, radians
finaldf1=[]
R = 6373.0
for latsub,longsub,suburb,postcode in zip(df1.Latitude, df1.Longitude,df1.Suburb,df1.Postcode): 
    for lathosp,longhosp,clinic,add,state in zip(df_center.latitude, df_center.longitude,df_center.Clinic_Name,df_center.Address,df_center.State):
        lat2 = radians(lathosp)
        lon2 = radians(longhosp)
        lat1 = radians(latsub)
        lon1 = radians(longsub)
        dlon = lon2 - lon1
        dlat = lat2 - lat1
        

        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))

        distance = R * c
        
        d = {
        'Clinic Name' : clinic, 
        'Address':add,
        'Suburb' : suburb,
        'Potcode_suburb' : postcode,
        'Latitude_suburb': latsub,
        'Longitude_suburb': longsub,
        'distance' : distance,
        'State': state
        }
        
        finaldf1.append(d)
finaldf1 = pd.DataFrame(finaldf1)
```

```python
finaldf1.head(5)
```

```python

g = finaldf1.groupby(["Suburb"]).apply(lambda x: x.sort_values(["distance"], ascending = True)).reset_index(drop=True)
df_suburbsdist=g.groupby('Suburb').head(5)
```
